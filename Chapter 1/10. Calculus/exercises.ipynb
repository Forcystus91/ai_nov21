```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative
```

1. Plot the absolute value of $x$ and its derivative. Choose an interval to have a nice looking shape.

```python
def abs(x):
    return np.abs(x)
```

```python
x= np.linspace(-10,10,100)
y = abs(x)

abs_derivative = derivative(abs,x,dx=1e-6)

plt.figure(figsize=((14,8)))
ax = plt.plot(x,y)
ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
plt.plot(x,abs_derivative)
plt.legend(['|x|','derivative of |x|'])

plt.show()
```

2. Plot the function $x\log{x}$ and its derivative.

```python
def log(x):
    return np.log(x)
```

```python
x = np.linspace(-10,10,100)
y = log(x)
log_derivative = derivative(log,x,1e-6)

plt.figure(figsize=(14,8))
plt.plot(x,y)
plt.plot(x,log_derivative)
plt.legend(['log(x)','derivative of log'])
plt.show
```

3. Plot the function $e^{x}\log{x}+5$ and its derivative.

```python
def elog_plus_5(x):
    return (np.exp(x)*log(x)) + 5
```

```python
x = np.linspace(-10,10,100)
y = elog_plus_5(x)
elog_derivative = derivative(elog_plus_5,x,1e-6)

plt.plot(x,y)
plt.plot(x, elog_derivative)
```

4. Plot the function $\sin(\frac{1}{2}x)+2$ in the interval $[0, 2\pi]$. Plot in the same graph both the function and the tangent line to the points which the function has derivative = 0.

```python
def sin(x):
    return np.sin(x)
def cos(x):
    return np.cos(x)
```

```python
x = np.linspace(0,2*np.pi, 1000)
y = sin(x/2) + 2

plt.figure(figsize=((14,8)))
plt.plot(x,y)
plt.plot(x,np.linspace(3,3, 1000))
plt.plot(x, cos(x/2))
plt.plot(x,np.linspace(1,1, 1000))
plt.legend(['$\sin(\\frac{x}{2})+2$','tan','$\cos(\\frac{x}{2})$'])


plt.show()
```
