1. Can you make the algorithm to be implemented into function gradient_descent(function, X, max_it)?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative
import matplotlib.cm as cm
```

```python
def gradient_descent(function, X, max_it):
    x=X
    index = np.random.randint(0,len(x))
    point = x[index]
    derivative_in_point = derivative(function, point)
    it = 0
    points = [point]
    threshold = 0.01

    while np.abs(derivative_in_point) > threshold and it < max_it:
        if derivative_in_point < 0:
            x = x[index:]
            index = np.random.randint(0,len(x))
            point = x[index]
            points.append(point)
            derivative_in_point = derivative(function,point)
        else:
            x = x[0:index+1]
            index = np.random.randint(0,len(x))
            point = x[index]
            points.append(point)
            derivative_in_point = derivative(function,point)
        it +=1
    return points
```
