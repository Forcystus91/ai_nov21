```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

page = requests.get('https://weather.com/weather/tenday/l/69bedc6a5b6e977993fb3e5344e3c06d8bc36a1fb6754c3ddfb5310a3c6d6c87')
soup = BeautifulSoup(page.content,'html.parser')
```

```python
# scraping site for data

day = soup.find_all('h2',class_="DetailsSummary--daypartName--2FBp2")
desc = soup.find_all('p',class_="DailyContent--narrative--hplRl")
temperature = soup.find_all('span', class_="DailyContent--temp--3d4dn")
```

```python
# creating lists
day_list = []
desc_day_list = []
desc_night_list = []
temp_low = []
temp_high = []

for i in range(10):
    day_list.append(day[i].text)
    for j in range(len(day_list[i])):
        day_list[i] = ''.join([k for k in day_list[i] if not k.isdigit()])  # removes digits from string
        day_list[i] = day_list[i].strip()   # removes blank space from string

for i in range(20):
    if i % 2 == 1:
        desc_night_list.append(desc[i].text)
        temp_low.append(int(temperature[i].text.strip('°')))
    else:
        desc_day_list.append(desc[i].text)
        temp_high.append(int(temperature[i].text.strip('°')))
```

```python
# converting temperature to celsius
def farhenheit_to_celsius(farh):
    celsius = (farh - 32) * (5/9)
    return f'{int(celsius)}\N{DEGREE SIGN}C'

highs_in_celsius = []
lows_in_celsius = []
for i in range(len(temp_high)):
    highs_in_celsius.append(farhenheit_to_celsius(temp_high[i]))
    lows_in_celsius.append(farhenheit_to_celsius(temp_low[i]))
```

```python
dates_list = pd.date_range('2021-12-08','2021-12-17')


forecast = pd.DataFrame({
    'Day':day_list,
    'Date':dates_list,
    'Daytime':desc_day_list,
    'Nighttime':desc_night_list,
    'High':highs_in_celsius,
    'Low':lows_in_celsius
})
forecast
```
