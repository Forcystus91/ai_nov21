Extract the 5 days, day and night (10 entries) of weather from Los Angeles in weather.gov.

write code to target the right elements of the text

For each day, you need to store:

The day of the week (Tuesday)

The date (28/09/2020)

A short description of the conditions (Clear early then increasing cloudiness after midnight. Low 41F. Winds light and variable)

The temperature low and high, with a function of your own to convert into Celsius

Save all of this into a Pandas dataframe

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime

five_day_list = ['Tuesday','Wednesday','Thursday','Friday','Saturday']
dates_list = pd.date_range('2021-12-07','2021-12-11')

page = requests.get('https://forecast.weather.gov/MapClick.php?lat=34.053570000000036&lon=-118.24544999999995#.Ya9chL3MKUm') 

soup = BeautifulSoup(page.content, 'html.parser')

div = soup.find_all('div', id='detailed-forecast-body')
string = div[0].text
list = []

with open('7_day_forecast.txt','w+') as fp:
    empty_string = ''
    for i in range(len(string)):
        if ord(string[i]) != 46:
            empty_string = empty_string + string[i]
        else:
            list.append(empty_string)
            empty_string = ''
    for j in range(3,len(list)-5):
        fp.write(f"{list[j].strip()}\n")

with open('7_day_forecast.txt') as fp:
    details = []
    line = fp.readline().strip()
    while line:
        details.append(line)
        line = fp.readline().strip()

def list_to_string(list):
    tmp_string = ''
    for i in range(len(list)):
        tmp_string = tmp_string + list[i]
    return tmp_string

tuesday = list_to_string(details[:6])
wednesday = list_to_string(details[6:13])
thursday = list_to_string(details[13:18])
friday = list_to_string(details[18:20])
saturday = list_to_string(details[20:])[8:]

summary = [tuesday,wednesday,thursday,friday,saturday]



high_temp = soup.find_all('p',class_='temp temp-high')
low_temp = soup.find_all('p',class_='temp temp-low')
five_day_highs = []
five_day_lows = []
for i in range(4):
    five_day_highs.append(high_temp[i].text)
for j in range(1,4):
    five_day_lows.append(low_temp[i].text)

for i in range(5-len(five_day_lows)):
    five_day_lows.append('NaN')
for i in range(5-len(five_day_highs)):
    five_day_highs.append('NaN')


forecast = pd.DataFrame({
    'Days of the Week':five_day_list,
    'Dates':dates_list,
    'Description':summary,
    'High':five_day_highs,
    'Low':five_day_lows
})

forecast



```
