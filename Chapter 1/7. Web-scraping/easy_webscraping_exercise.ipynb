Extract the 5 days, day and night (10 entries) of weather from Los Angeles in weather.gov.

write code to target the right elements of the text

For each day, you need to store:

The day of the week (Tuesday)

The date (28/09/2020)

A short description of the conditions (Clear early then increasing cloudiness after midnight. Low 41F. Winds light and variable)

The temperature low and high, with a function of your own to convert into Celsius

Save all of this into a Pandas dataframe

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

five_day_list = ['Tuesday','Wednesday','Thursday','Friday','Saturday']
dates_list = pd.date_range('2021-12-07','2021-12-11')


page = requests.get('https://forecast.weather.gov/MapClick.php?lat=34.053570000000036&lon=-118.24544999999995#.Ya9qCL3MKUn') 
soup = BeautifulSoup(page.content, 'html.parser')

div = soup.find_all('div', id='seven-day-forecast-container')
text = div[0].text.strip()

# Creating a list of the websites content
string = ''
tmp_list = []
for i in range(len(text)):
    if ord(text[i]) != 10:
        string = string + text[i]
    else:
        if string != '':
            tmp_list.append(string)
            string = ''
else:
    tmp_list.append(string)


# Storing temperature values in a list
temp = []
for i in range(1,len(tmp_list),2):
    temp.append(int(tmp_list[i][-5:-3]))


#   converting farhenheit to Celsius
def farhenheit_to_celsius(farhenheit):
    celsius = (farhenheit - 32) * (5/9)
    return f'{int(celsius)}\N{DEGREE SIGN}'

farhenheit_temp_highs = []
farhenheit_temp_lows = []
for i in range(0,len(temp),2):
    farhenheit_temp_highs.append(farhenheit_to_celsius(temp[i]))

for j in range(1,len(temp),2):
    farhenheit_temp_lows.append(farhenheit_to_celsius(temp[j]))

farhenheit_temp_lows.append('NaN')
farhenheit_temp_lows

#   Creating list of descriptions for each day
description_odd = []
for i in range(0,len(tmp_list),4):
    description_odd.append(tmp_list[i+1][:-11])

description_even = []
for j in range(1,len(tmp_list)-1,4):
    description_even.append(tmp_list[j+2][:-10])
description_even.append('...')

descriptions = []
for i in range(len(description_even)):
    descriptions.append(description_odd[i].strip() +', then ' + description_even[i].strip())


# Creating DataFrame
forecast = pd.DataFrame({
    'Day of the Week':five_day_list,
    'Date':dates_list,
    'Description':descriptions,
    'High':farhenheit_temp_highs,
    'Low':farhenheit_temp_lows
})
forecast
```
