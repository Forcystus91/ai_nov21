### Exercise 2: Get the Total profit of all months and show a line plot with the following Style properties:

- X label name = Month Number
- Y label name = Sold units number

### Exercise 8: Calculate the total sale data for last year for each product and show it using a Pie chart.
Note: In the Pie chart you should display the Number of units sold per year for each product in percentage.

The Pie chart should look like this:


```python
labels = ['Face Cream','Face Wash','Toothpaste','Bathing Soap','Shampoo','Moisturizer']
values = [data['facecream'].sum(),data['facewash'].sum(),data['toothpaste'].sum(),data['bathingsoap'].sum(),data['shampoo'].sum(),data['moisturizer'].sum()]

plt.figure(figsize=(14,10))
plt.pie(values, labels=labels,autopct='%1.1f%%',textprops={'fontsize':14})
plt.legend(loc='lower right')
plt.title('Sales Data')

plt.show()
```

###Â Exercise 9: Read the Bathing soap facewash of all months and display it using the Subplot.
The Subplot should look like this:


```python
plt.figure(figsize=(14,10))

plt.figure(1)   # The first figure
plt.subplot(211)    # the first subplot
plt.plot(x,data['bathingsoap'])
plt.title('Sales data of bathsoap')
plt.subplot(212)    # next subplot, current subplot under figure 1
plt.plot(x,data['facewash'])
plt.title('Sales data of facewash')


plt.subplot(211)    #makes subplot(211) in figure1 current
plt.show()
```
